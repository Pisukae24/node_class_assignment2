```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    
    def __init__(self):
        self.head = None

    def add_node_at_end(self, data):
       
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        
        temp = self.head
        while temp.next is not None:
            temp = temp.next
        temp.next = new_node

    def print_list(self):
        
        if self.head is None:
            print("The list is empty.")
            return

        temp = self.head
        elements = []
        while temp is not None:
            elements.append(str(temp.data))
            temp = temp.next
        print(" -> ".join(elements))

    def delete_node_by_index(self, n):
        
        if self.head is None:
            raise Exception("Can't delete from an empty list.")
        
        if n <= 0:
            raise Exception("Index must be 1 or greater.")
        
        if n == 1:
            self.head = self.head.next
            return

        temp = self.head
        count = 1
        while temp is not None and count < n-1:
            temp = temp.next
            count += 1

        if temp is None or temp.next is None:
            raise Exception("Index out of range.")
        
        temp.next = temp.next.next


if __name__ == "__main__":
    
    llist = LinkedList()
    llist.add_node_at_end(10)
    llist.add_node_at_end(20)
    llist.add_node_at_end(30)
    llist.add_node_at_end(40)

    print("List after adding nodes:")
    llist.print_list()

   
    try:
        llist.delete_node_by_index(2)
        print("List after removing 2nd node:")
        llist.print_list()
    except Exception as e:
        print("Error :", e)

    
    try:
        empty = LinkedList()
        empty.delete_node_by_index(1)
    except Exception as e:
        print("Error :", e)

    
    try:
        llist.delete_node_by_index(10)
    except Exception as e:
        print("Error :", e)
```
